```python
#Realiza el an√°lisis de las variables usando diagramas de cajas y bigotes, histogramas y mapas de calor.

#----------Bigotes----------

#Average Price
boxplot = data.boxplot(column=['AveragePrice'])
boxplot.plot()
plt.show()

#Total Volume
boxplot = data.boxplot(column=['Total Volume'])
boxplot.plot()
plt.show()

#4046
boxplot = data.boxplot(column=['4046'])
boxplot.plot()
plt.show()

#4225
boxplot = data.boxplot(column=['4225'])
boxplot.plot()
plt.show()

#4770
boxplot = data.boxplot(column=['4770'])
boxplot.plot()
plt.show()

#Total Bags
boxplot = data.boxplot(column=['Total Bags'])
boxplot.plot()
plt.show()

#Small Bags
boxplot = data.boxplot(column=['Small Bags'])
boxplot.plot()
plt.show()

#Large Bags
boxplot = data.boxplot(column=['Large Bags'])
boxplot.plot()
plt.show()

#XLarge Bags
boxplot = data.boxplot(column=['XLarge Bags'])
boxplot.plot()
plt.show()

#year
boxplot = data.boxplot(column=['year'])
boxplot.plot()
plt.show()

```

```python
#----------Histograma----------
##Op 1:
plt.style.use('ggplot')
plt.hist(data['AveragePrice'], bins=10)
plt.show()

##Op2
plt.style.use('ggplot')

density = stats.gaussian_kde(data['AveragePrice'])
n, x, _ =plt.hist(data['AveragePrice'], bins=np.linspace(0, 3, 50), 
                   histtype=u'step', density=True)
plt.plot(x, density(x))
plt.show()
```

```python
#----------Mapa de calor----------
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(data.corr(), center=0, cmap= 'Reds')
```
